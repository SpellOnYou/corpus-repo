```
def hwp_version_test(file_list):
  success_list = []
  for idx, name in enumerate(file_list):
    FILE = name
    temp = !/usr/local/bin/hwp5proc version $FILE
    if 'Not' in temp[-1]: #실제 버전이 안 받는 파일은 compound binary file이라 인식하지 못하기 때문에 위와 같은 실행 결과가 출력된다.
      print(idx, FILE, "FAIL", temp)
    else:
      vs = !/usr/local/bin/hwp5proc version $FILE
      print(idx, FILE, "PASS", vs)
      success_list.append(FILE)
  return(success_list)
```

```
success_list = hwp_version_test(os.listdir('.'))
```

```
def save_txt(per_list): #per_list : version 통과를 받은 한글파일 리스트
  for idx, file_name in enumerate(per_list):
    a = time.time() #txt 파일로 변환하기 까지 걸린 시간을 체크하기 위함
    path_dir = '/content/godrive/My Drive/corpus-repo/corpus/hwp_to_txt'
    dst_dir = '/content/godrive/My Drive/corpus-repo/corpus'
    # 디렉토리가 없을시 생성
    if not os.path.exists(dst_dir):
      os.makedirs(dst_dir)
      
    #쉘 명령어 수행하기 위함
    TO = os.path.join(dst_dir, file_name )
    FROM = os.path.join(path_dir+file_name)
    
    !/usr/local/bin/hwp5txt $FROM > $TO
    
    print(idx, 'file:', os.path.basename(TO), os.path.isfile(TO), str(time.time()-a)+'초')
```
